# Analisis Deskriptif & Visualisasi Data Coffee Shop

## 1. Import Library

```python
import numpy as np                  # untuk komputasi numerik
import pandas as pd                 # untuk manipulasi dan pengolahan data tabular
import matplotlib.pyplot as plt     # untuk visualisasi data dasar
import seaborn as sns               # untuk visualisasi data lanjutan

# mengatur format output dari pandas (opsional)
pd.options.display.float_format = '{:,.2f}'.format
```

## 2.1 Contoh sederhana

```python
# contoh data nilai mahasiswa
nilai = [80, 76, 84, 68, 45, 72, 67, 98, 82, 89, 42, 95, 61, 73, 88, 96, 77, 64, 91, 87]

# Menghitung statistik dasar
mean_nilai = np.mean(nilai)                     # Rata-rata
median_nilai = np.median(nilai)                 # Median

std_nilai = np.std(nilai)                       # Standar Deviasi
var_nilai = np.var(nilai)                       # Varians

min_nilai = np.min(nilai)                       # Nilai Minimum
max_nilai = np.max(nilai)                       # Nilai Maksimum
range_nilai = max_nilai - min_nilai             # Rentang (Range)

q1_nilai = np.percentile(nilai, 25)             # Kuartil 1
q3_nilai = np.percentile(nilai, 75)             # Kuartil 3
iqr_nilai = q3_nilai - q1_nilai                 # Jangkauan Interkuartil

skew_nilai = pd.Series(nilai).skew()            # Skewness
kurtosis_nilai = pd.Series(nilai).kurtosis()    # Kurtosis

# Menampilkan hasil statistik
print("Statistik Dasar Nilai:")
print(f"Rata-rata: {mean_nilai:.2f}")
print(f"Median: {median_nilai:.2f}")
print(f"Standar Deviasi: {std_nilai:.2f}")
print(f"Varians: {var_nilai:.2f}")
print(f"Nilai Minimum: {min_nilai:.2f}")
print(f"Nilai Maksimum: {max_nilai:.2f}")
print(f"Kuartil 1 (Q1): {q1_nilai:.2f}")
print(f"Kuartil 3 (Q3): {q3_nilai:.2f}")
print(f"Rentang Interkuartil (IQR): {iqr_nilai:.2f}")
print(f"Rentang: {range_nilai:.2f}")
```

## 2. 2 Contoh Menggunakan Dataset
Data diambil dari file `dataset_coffeeshop.csv`.

```python
# Membaca dataset dari file CSV (lokal)
# data = pd.read_csv('dataset_coffeeshop.csv')

# Membaca dataset dari URL
data = pd.read_csv('https://github.com/sainsdataid/dataset/raw/refs/heads/main/dataset_coffeeshop.csv')

# Menampilkan 5 baris pertama dari dataset
data.head()
```

## 3. Analisis Deskriptif Univariat
Pada bagian ini kita membahas statistik deskriptif:
- **Ukuran Pemusatan:** Rata-rata (mean), median, modus (mode)
- **Ukuran Penyebaran:** range, varians, standar deviasi, IQR
- **Ukuran Lokasi:** kuartil, persentil
- **Distribusi frekuensi untuk variabel kategorik**

### 3.1 Menampilkan ringkasan statistik deskriptif dari dataset

```python
# Menampilkan Ringkasan statistik deskriptif dari dataset
data.describe()
```

### 3.2 Menampilkan Statistik Deskriptif Kolom Numerik

```python
# Menampilkan Ringkasan statistik deskriptif dari kolom 'spending_mil'
data["spending_mil"].describe()
```

```python
# statistik deskriptif untuk kolom numerik tertentu

mean_spending = data["spending_mil"].mean()            # Rata-rata
median_spending = data["spending_mil"].median()        # Median

std_spending = data["spending_mil"].std()              # Standar Deviasi
var_spending = data["spending_mil"].var()              # Varians

min_spending = data["spending_mil"].min()              # Nilai Minimum
max_spending = data["spending_mil"].max()              # Nilai Maksimum
range_spending = max_spending - min_spending           # Rentang (Range)

q1_spending = data["spending_mil"].quantile(0.25)      # Kuartil 1
q3_spending = data["spending_mil"].quantile(0.75)      # Kuartil 3
iqr_spending = q3_spending - q1_spending               # Jangkauan Interkuartil

skew_spending = data["spending_mil"].skew()            # Skewness
kurtosis_spending = data["spending_mil"].kurtosis()    # Kurtosis


# Menampilkan hasil statistik deskriptif untuk kolom 'spending_mil'
print("\nStatistik Deskriptif untuk Kolom 'spending_mil':")
print(f"Rata-rata: {mean_spending:.2f}")
print(f"Median: {median_spending:.2f}")
print(f"Standar Deviasi: {std_spending:.2f}")
print(f"Varians: {var_spending:.2f}")
print(f"Nilai Minimum: {min_spending:.2f}")
print(f"Nilai Maksimum: {max_spending:.2f}")
print(f"Kuartil 1 (Q1): {q1_spending:.2f}")
print(f"Kuartil 3 (Q3): {q3_spending:.2f}")
print(f"Rentang Interkuartil (IQR): {iqr_spending:.2f}")
print(f"Rentang: {range_spending:.2f}")
print(f"Skewness: {skew_spending:.2f}")
print(f"Kurtosis: {kurtosis_spending:.2f}")
```

### 3.3 Menampilkan Statistik Variabel Kategorik

```python
# fungsi bantuan untuk membuat tabel frekuensi
def freq_table(series):
    t = series.value_counts().to_frame('count')
    t['percent'] = (t['count']/len(series)*100).round(2)
    return t

freq_gender = freq_table(data['gender'])
display(freq_gender)

freq_city = freq_table(data['city'])
display(freq_city)
```

## 4. Analisis Bivariat
### 4.1 Kategorik vs Numerik  *Contoh:* Spending berdasarkan Gender


```python
 # menampilkan rata-rata pengeluaran berdasarkan gender
data.groupby('gender')['spending_mil'].mean().round(2)
```

### 4.2 Numerik vs Numerik  *Contoh:* Korelasi Spending dan Frekuensi Kunjungan

```python
 # menghitung korelasi antara pengeluaran dan frekuensi kunjungan
data['spending_mil'].corr(data['frequency_visit'])
```

### 4.3 Kategorik vs Kategorik  *Contoh:* Gender vs City

```python
   # menampilkan tabel kontingen antara gender dan city
ct = pd.crosstab(data['gender'], data['city'])
ct
```

## 5. Visualisasi Data
Visualisasi menunjukkan pola distribusi dan hubungan antar variabel.

```python
# Mengatur tema seaborn tanpa grid
sns.set_theme(style="whitegrid")
sns.set_style("white")
```

### 5.1 Histogram Variabel Numerik

```python
# Membuat histogram untuk kolom 'spending_mil'
plt.figure(figsize=(6,4))
sns.histplot(data['spending_mil'], bins=30, kde=True)
plt.title('Histogram Spending (Juta)')
plt.show()

# Membuat histogram untuk kolom 'frequency_visit'
plt.figure(figsize=(6,4))
sns.histplot(data['frequency_visit'], bins=20, color='red')
plt.title('Histogram Frekuensi Kunjungan')
plt.show()
```

### 5.2 Boxplot

```python
# Membuat boxplot untuk kolom 'spending_mil'
plt.figure(figsize=(6,3))
sns.boxplot(x=data['spending_mil'])
plt.title('Boxplot Spending')
plt.show()

# Membuat boxplot untuk kolom 'frequency_visit' berdasarkan 'city'
plt.figure(figsize=(6,3))
sns.boxplot(x='gender', y='frequency_visit', data=data,
            hue='gender', width= 0.5,
            linewidth=2, linecolor="#022F21",
            flierprops={'marker':'d', 'markerfacecolor':'darkblue', 'markersize':8})
plt.title('Boxplot Frekuensi Kunjungan berdasarkan Gender')
plt.show()
```

### 5.3 Barplot (Kategorik vs Numerik)

```python
# Membuat barplot rata-rata spending per kota
plt.figure(figsize=(7,4))
sns.barplot(x='city', y='spending_mil', data=data, color="#496AD5", edgecolor='black', linewidth=0.75, width=0.7, errorbar=None)
plt.title('Rata-rata Spending per Kota', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Kota', fontsize=12, fontweight='bold')
plt.ylabel('Spending (Juta Rupiah)', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 5.4 Scatter Plot (Numerik vs Numerik)

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=data, x='frequency_visit', y='spending_mil', alpha=0.6)
plt.title('Spending vs Frekuensi Kunjungan')
plt.show()
```

### 5.5 Heatmap (Kategorik vs Kategorik)

```python
plt.figure(figsize=(7,4))
sns.heatmap(ct, annot=True, cmap='Greens', fmt='d')
plt.title('Gender vs City')
plt.show()

# Membuat heatmap rata-rata spending berdasarkan gender dan city
plt.figure(figsize=(7,4))
pivot_spending = data.pivot_table(values='spending_mil', index='gender', columns='city', aggfunc='mean')
sns.heatmap(pivot_spending, annot=True, fmt='.2f', cmap='YlOrRd')
plt.title('Rata-rata Spending per Gender dan City (Juta)', fontsize=12, fontweight='bold', pad=10)
plt.ylabel('Gender', fontsize=11, fontweight='bold')
plt.xlabel('City', fontsize=11, fontweight='bold')
plt.tight_layout()
plt.show()
```
